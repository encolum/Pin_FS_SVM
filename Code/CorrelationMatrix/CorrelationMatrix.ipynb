```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu gốc
df = pd.read_csv(r"wdbc.data.txt", header=None)

# Tách các biến (bỏ cột đầu tiên là ID)
X = df.iloc[:, 1:]

# Đổi tên cột thứ hai thành 'Target' và chuyển đổi giá trị
X = X.rename(columns={1: 'Target'})
X['Target'] = np.where(X['Target'] == 'M', 1, -1)

# Đổi tên tất cả các cột (trừ 'Target') bằng cách giảm giá trị xuống 1
new_columns = {'Target': 'Target'}
for col in X.columns:
    if col != 'Target':
        new_columns[col] = f'Feature_{col - 1}'

X = X.rename(columns=new_columns)

# Tính ma trận tương quan
corr_matrix = X.corr()

# Vẽ heatmap cho ma trận tương quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu')

plt.title('Correlation Matrix')
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Lấy 10 đặc trưng có ảnh hưởng lớn nhất (bỏ qua 'Target' vì nó luôn có tương quan 1.0)
top_10_features = target_correlations[1:11]

# lấy ra 50% đặc trưng có tương quan thấp nhất so với Target
low_features = target_correlations[21:35]

print("10 đặc trưng có ảnh hưởng lớn nhất đến biến mục tiêu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
print("10 đặc trưng có ảnh hưởng thấp nhất đến biến mục tiêu:")
for feature, correlation in low_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Đọc dữ liệu gốc
df = pd.read_csv(r"ionosphere.data", header=None)

# Đổi tên cột cuối cùng thành 'Target' và chuyển đổi giá trị

# Đổi tên cột cuối cùng thành 'Target' và chuyển đổi giá trị
df = df.rename(columns={df.columns[-1]: 'Target'})
df['Target'] = np.where(df['Target'] == 'g', 1, -1)

# Đổi tên các cột còn lại
new_columns = {col: f'Feature_{col + 1}' for col in df.columns if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyển cột 'Target' lên đầu DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]

# Tính ma trận tương quan
corr_matrix = df.corr()

# Vẽ heatmap cho ma trận tương quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu', cbar=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Lấy 10 đặc trưng có ảnh hưởng lớn nhất (bỏ qua 'Target' vì nó luôn có tương quan 1.0)
top_10_features = target_correlations[1:11]

top_10_low_features = target_correlations[24:35]

print("10 đặc trưng có ảnh hưởng lớn nhất đến biến mục tiêu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
print("10 đặc trưng có ảnh hưởng thấp nhất đến biến mục tiêu:")
for feature, correlation in top_10_low_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu gốc
df = pd.read_csv(r"diabetes.csv")



# Đổi tên cột thứ hai thành 'Target' và chuyển đổi giá trị
df = df.rename(columns={df.columns[-1]: 'Target'})
df['Target'] = np.where(df['Target'] == 0, -1, 1)

new_columns = {col: f'Feature_{i+1}' for i, col in enumerate(df.columns) if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyển cột 'Target' lên đầu DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]


# Tính ma trận tương quan
corr_matrix = df.corr()

# Vẽ heatmap cho ma trận tương quan
plt.figure(figsize=(40, 30))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('Correlation Matrix')
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Lấy 10 đặc trưng có ảnh hưởng lớn nhất (bỏ qua 'Target' vì nó luôn có tương quan 1.0)
top_10_features = target_correlations[1:11]

print("10 đặc trưng có ảnh hưởng lớn nhất đến biến mục tiêu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Đọc dữ liệu gốc
df = pd.read_csv(r"Heart_disease_cleveland_new.csv")

# Đổi tên cột cuối cùng thành 'Target' và chuyển đổi giá trị

# Đổi tên cột cuối cùng thành 'Target' và chuyển đổi giá trị
df = df.rename(columns={'target': 'Target'})
df['Target'] = np.where(df['Target'] == 0, -1, 1)
# Đổi tên các cột còn lại
new_columns = {col: f'Feature_{i+1}' for i, col in enumerate(df.columns) if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyển cột 'Target' lên đầu DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
# df = df.iloc[1:,:]
# Tính ma trận tương quan
corr_matrix = df.corr()

# Vẽ heatmap cho ma trận tương quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu', cbar=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Lấy 10 đặc trưng có ảnh hưởng lớn nhất (bỏ qua 'Target' vì nó luôn có tương quan 1.0)
top_10_features = target_correlations[1:11]

print("10 đặc trưng có ảnh hưởng lớn nhất đến biến mục tiêu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
```
