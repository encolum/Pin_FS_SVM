# Wdbc 

Original

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# c d liu gc
df = pd.read_csv(r"wdbc.data.txt", header=None)

# Tch cc bin (b ct u tin l ID)
X = df.iloc[:, 1:]

# i tn ct th hai thnh 'Target' v chuyn i gi tr
X = X.rename(columns={1: 'Target'})
X['Target'] = np.where(X['Target'] == 'M', 1, -1)

# i tn tt c cc ct (tr 'Target') bng cch gim gi tr xung 1
new_columns = {'Target': 'Target'}
for col in X.columns:
    if col != 'Target':
        new_columns[col] = f'Feature_{col - 1}'

X = X.rename(columns=new_columns)

# Tnh ma trn tng quan
corr_matrix = X.corr()

# V heatmap cho ma trn tng quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu')

plt.title('Correlation Matrix')
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Ly 10 c trng c nh hng ln nht (b qua 'Target' v n lun c tng quan 1.0)
top_10_features = target_correlations[1:11]

# ly ra 50% c trng c tng quan thp nht so vi Target
low_features = target_correlations[21:35]

print("10 c trng c nh hng ln nht n bin mc tiu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
print("10 c trng c nh hng thp nht n bin mc tiu:")
for feature, correlation in low_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# c d liu gc
df = pd.read_csv(r"ionosphere.data", header=None)

# i tn ct cui cng thnh 'Target' v chuyn i gi tr

# i tn ct cui cng thnh 'Target' v chuyn i gi tr
df = df.rename(columns={df.columns[-1]: 'Target'})
df['Target'] = np.where(df['Target'] == 'g', 1, -1)

# i tn cc ct cn li
new_columns = {col: f'Feature_{col + 1}' for col in df.columns if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyn ct 'Target' ln u DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]

# Tnh ma trn tng quan
corr_matrix = df.corr()

# V heatmap cho ma trn tng quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu', cbar=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Ly 10 c trng c nh hng ln nht (b qua 'Target' v n lun c tng quan 1.0)
top_10_features = target_correlations[1:11]

top_10_low_features = target_correlations[24:35]

print("10 c trng c nh hng ln nht n bin mc tiu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
print("10 c trng c nh hng thp nht n bin mc tiu:")
for feature, correlation in top_10_low_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# c d liu gc
df = pd.read_csv(r"australia.txt", sep = ' ', header= None)


# i tn ct th hai thnh 'Target' v chuyn i gi tr
df = df.rename(columns={df.columns[-1]: 'Target'})
df['Target'] = np.where(df['Target'] == 0, -1, 1)

new_columns = {col: f'Feature_{i+1}' for i, col in enumerate(df.columns) if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyn ct 'Target' ln u DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]


# Tnh ma trn tng quan
corr_matrix = df.corr()

# V heatmap cho ma trn tng quan
plt.figure(figsize=(40, 30))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('Correlation Matrix')
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Ly 10 c trng c nh hng ln nht (b qua 'Target' v n lun c tng quan 1.0)
top_10_features = target_correlations[1:8]

print("10 c trng c nh hng ln nht n bin mc tiu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# c d liu gc
df = pd.read_csv(r"Heart_disease_cleveland_new.csv")

# i tn ct cui cng thnh 'Target' v chuyn i gi tr

# i tn ct cui cng thnh 'Target' v chuyn i gi tr
df = df.rename(columns={'target': 'Target'})
df['Target'] = np.where(df['Target'] == 0, -1, 1)
# i tn cc ct cn li
new_columns = {col: f'Feature_{i+1}' for i, col in enumerate(df.columns) if col != 'Target'}
new_columns['Target'] = 'Target'
df = df.rename(columns=new_columns)

# Di chuyn ct 'Target' ln u DataFrame
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
# df = df.iloc[1:,:]
# Tnh ma trn tng quan
corr_matrix = df.corr()

# V heatmap cho ma trn tng quan
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu', cbar=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
print('*'*100)
target_correlations = corr_matrix['Target'].abs().sort_values(ascending=False)

# Ly 10 c trng c nh hng ln nht (b qua 'Target' v n lun c tng quan 1.0)
top_10_features = target_correlations[1:11]

print("10 c trng c nh hng ln nht n bin mc tiu:")
for feature, correlation in top_10_features.items():
    print(f"{feature}: {correlation:.4f}")
```
