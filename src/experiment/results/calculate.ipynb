```python
import pandas as pd 
import glob 
import os
PATH = os.getcwd()  
while not os.path.basename(PATH).startswith('Optimal-Robust-Feature-Selection'):
    parent = os.path.dirname(PATH)
    if parent == PATH:  
        break
    PATH = parent
results_dir = os.path.join(PATH, 'src', 'experiment', 'results')
datasets = ['cleveland', 'diabetes', 'ionosphere', 'sonar', 'wdbc']
list = [os.path.join(results_dir, f'experiment_results_{dataset}.xlsx') for dataset in datasets]
```

```python
df_list = [pd.read_excel(file) for file in list]
```

```python
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
combined_df = combined_df.iloc[:,:8]
```

```python
combined_df
```

```python
avg_df = combined_df.groupby(['Model', 'Type of dataset']).mean(numeric_only=True).reset_index()
```
