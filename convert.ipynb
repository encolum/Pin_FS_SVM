```python
import pandas as pd 
import numpy as np
df = pd.read_excel(r'D:\Optimal-Robust-Feature-Selection-For-Support-Vector-Machines-With-Pinball-Loss\src\experiment\results\average_metrics_per_model_dataset.xlsx')
```

```python
df
```

```python
# Group by Model and multiply numeric values by 100
df_grouped = df.groupby('Model').apply(lambda group: group.copy())

# Function to multiply values in string format by 100
def multiply_string_values(value_str):
    if '±' in str(value_str):
        parts = str(value_str).split(' ± ')
        mean_val = float(parts[0]) * 100
        std_val = float(parts[1]) * 100
        return f"{mean_val:.2f} ± {std_val:.2f}"
    return value_str

# Apply the transformation to numeric columns (exclude Model and Type of dataset)
numeric_columns = ['Average Accuracy', 'Average AUC', 'Average F1 Score', 'Average G-Mean', ' Average Overlap Ratio']

for col in numeric_columns:
    df_grouped[col] = df_grouped[col].apply(multiply_string_values)

# Reset index to make it easier to work with (NGOÀI LOOP)
df_grouped = df_grouped.reset_index(drop=True)

# Define the desired order for Type of dataset (NGOÀI LOOP)
dataset_order = ['Not noise', 'Noise', 'Outlier', 'Noise + Outlier']

# Create a categorical column with the specified order (NGOÀI LOOP)
df_grouped['Type of dataset'] = pd.Categorical(df_grouped['Type of dataset'], 
                                               categories=dataset_order, 
                                               ordered=True)

# Sort by Model and then by Type of dataset (NGOÀI LOOP)
df_grouped = df_grouped.sort_values(['Model', 'Type of dataset']).reset_index(drop=True)

df_grouped
```

```python
df_grouped.to_excel(r'D:\Optimal-Robust-Feature-Selection-For-Support-Vector-Machines-With-Pinball-Loss\src\experiment\results\average_metrics_per_model_dataset_formatted.xlsx', index=False)
```
